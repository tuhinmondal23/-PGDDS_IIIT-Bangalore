# Missing/Null Values Treatment

These are the columns with Missing Values, will treat them one by one:
    
1.OCCUPATION_TYPE -- 31.34%

2.AMT_REQ_CREDIT_BUREAU_YEAR  --  13.5016%

3.NAME_TYPE_SUITE  --  0.4201%

6.AMT_GOODS_PRICE  --  0.0904%

7.AMT_ANNUITY  --  0.039%

8.CNT_FAM_MEMBERS  --  0.0007%

```python
# 1. OCCUPATION_TYPE is a categorical column
#    It's having moderately high percentage of Missing Values
#    So will replace the missing values with "Missing" lebel

app_data["OCCUPATION_TYPE"].fillna("Missing",inplace=True)
```

```python
# Checking if any missing values are there 

app_data["OCCUPATION_TYPE"].isnull().sum()
```

```python
# 1. AMT_REQ_CREDIT_BUREAU_YEAR - Numeric column
#    will check for outliers and will impute the missing values with Median/Mean accordingly
#    Next change the datatype to Int from Float

sns.boxplot(app_data["AMT_REQ_CREDIT_BUREAU_YEAR"])
plt.show()
```

```python
# as the "AMT_REQ_CREDIT_BUREAU_YEAR" column has outliers, will replace the Null values with Median value

app_data["AMT_REQ_CREDIT_BUREAU_YEAR"].fillna(app_data["AMT_REQ_CREDIT_BUREAU_YEAR"].median(),inplace=True)
```

```python
# Checking if any missing values are there 

app_data["AMT_REQ_CREDIT_BUREAU_YEAR"].isnull().sum()
```

```python
# 3.NAME_TYPE_SUITE - Categorical column
# As the missing value percentage is quite low, will impute the missing values with Mode()

app_data["NAME_TYPE_SUITE"].fillna(app_data["NAME_TYPE_SUITE"].mode()[0],inplace=True)
```

```python
# Checking if any missing values are there 

app_data["NAME_TYPE_SUITE"].isnull().sum()
```

```python
# 4. DEF_60_CNT_SOCIAL_CIRCLE -- Numeric column
# 5. OBS_60_CNT_SOCIAL_CIRCLE -- Numeric column
# 6. AMT_GOODS_PRICE -- Numeric column
# 7. AMT_ANNUITY -- Numeric column
# 8. CNT_FAM_MEMBERS -- Numeric column

# will check for outliers and will impute the missing values with Median/Mean accordingly

from matplotlib.pyplot import figure

figure(figsize=(20,10))
plt.subplot(2,3,1)
sns.boxplot(app_data["DEF_60_CNT_SOCIAL_CIRCLE"])
plt.subplot(2,3,2)
sns.boxplot(app_data["OBS_60_CNT_SOCIAL_CIRCLE"])
plt.subplot(2,3,3)
sns.boxplot(app_data["AMT_GOODS_PRICE"])
plt.subplot(2,3,4)
sns.boxplot(app_data["AMT_ANNUITY"])
plt.subplot(2,3,5)
sns.boxplot(app_data["CNT_FAM_MEMBERS"])
plt.show()
```

```python
# as the columns have outliers, will replace the Null values with Median value

app_data["DEF_60_CNT_SOCIAL_CIRCLE"].fillna(app_data["DEF_60_CNT_SOCIAL_CIRCLE"].median(),inplace=True)
app_data["OBS_60_CNT_SOCIAL_CIRCLE"].fillna(app_data["OBS_60_CNT_SOCIAL_CIRCLE"].median(),inplace=True)
app_data["AMT_GOODS_PRICE"].fillna(app_data["AMT_GOODS_PRICE"].median(),inplace=True)
app_data["AMT_ANNUITY"].fillna(app_data["AMT_ANNUITY"].median(),inplace=True)
app_data["CNT_FAM_MEMBERS"].fillna(app_data["CNT_FAM_MEMBERS"].median(),inplace=True)
```

```python
# Checking if any missing values are there 

app_data[["DEF_60_CNT_SOCIAL_CIRCLE","OBS_60_CNT_SOCIAL_CIRCLE","AMT_GOODS_PRICE","AMT_ANNUITY","CNT_FAM_MEMBERS"]].isnull().sum()
```
